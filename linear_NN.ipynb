```python
import numpy as np
import tensorflow
from tensorflow import keras
```

```python
model= keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
x= np.array([1,3,5,7,9,11,13,15], dtype=float)
y= np.array([69,67,65,63,61,59,57,55],dtype=float)
```

```python
model.compile(loss="mean_squared_error",optimizer="sgd")
```

```python
model.fit(x,y,epochs=1500)
```

```python
z=np.array([17,19,21,23,25,27,29],dtype=float)
```

```python
model.predict(z)
```
